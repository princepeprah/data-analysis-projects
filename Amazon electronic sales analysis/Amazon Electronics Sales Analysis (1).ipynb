## Visualizations

```python
# Ratings distribution 
sns.countplot(x='rating', data=df)
```

```python
#Sales by years

plt.figure(figsize=(10,5))
sns.countplot(x='year', data=df)
```

```python
# Highest Sales by brands

plt.figure(figsize=(10,5))
sns.countplot(x='brand', data=df, order=df['brand'].value_counts().iloc[1:10].index)
```

Logitech recorded the highest number of sales followed by Sony with Pyle and Etre Jeune recording the least

```python
# Lowest sales by brands

plt.figure(figsize=(15,5))
sns.countplot(x='brand', data=df, order=df['brand'].value_counts().iloc[-10:].index)
```

Koolertron recorded the lowest sales followed by Duragadget

```python
# Highest sales in 2015 by brands

plt.figure(figsize=(12,6))
sns.countplot(x='brand', data=df[df['year'] == 2015], order=df['brand'].value_counts().iloc[1:10].index)
```

Mpow recorded the highest number of sales in 2015

```python
# Highest sales in 2016 by brands

plt.figure(figsize=(12,6))
sns.countplot(x='brand', data=df[df['year'] == 2016], order=df['brand'].value_counts().iloc[1:10].index)
```

Logitech recorded the highest number of sales in 2016

```python
# Highest sales in 2017 by brands - Logitech

plt.figure(figsize=(12,6))
sns.countplot(x='brand', data=df[df['year'] == 2017], order=df['brand'].value_counts().iloc[1:10].index)
```

```python
# Month with the highest sales - January

sns.countplot(x='month', data=df)
```

```python
# Category with the highest sales - Computers & Accessories

plt.figure(figsize=(20,6))
sns.countplot(x='category', data=df[df['month'] == 1], order=df['category'].value_counts().iloc[1:10].index)
```

```python
# Category with the least sales - Security & Surveillance

plt.figure(figsize=(20,6))
sns.countplot(x='category', data=df, order=df['category'].value_counts().iloc[-10:].index)
```

```python
plt.figure(figsize=(10,6))
df.groupby('category')['rating'].count().sort_values(ascending=False).head(10).plot(kind='pie')
```
